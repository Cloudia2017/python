```python
import sys
import importlib
importlib.reload(sys)

import urllib.request
from bs4 import BeautifulSoup

if __name__ == '__main__':
    url = 'http://top.17k.com/top/top100/06_vipclick/06_vipclick_cnl_mm_top_100_pc.html'
    head = {}
    head['User-Agent'] = 'Mozilla/5.0 (Linux; Android 4.1.1; Nexus 7 Build/JRO03D) AppleWebKit/535.19 (KHTML, like Gecko) Chrome/18.0.1025.166  Safari/535.19'
    req = urllib.request.Request(url, headers = head)
    response=urllib.request.urlopen(req)
    html=response.read()  
    soup = BeautifulSoup(html, 'lxml')
    soup_texts = soup.find('div', class_= 'TYPE')
    links = soup_texts.find_all('a', class_='red')
    for link in links[94:100]:
        if link!='\n':
            print(link.text + ':  ', link.get('href').replace('book','list'))
            url2 = link.get('href').replace('book','list')
            req2 = urllib.request.Request(url2, headers = head)
            response2=urllib.request.urlopen(req2)
            html2=response2.read()
            soup2 = BeautifulSoup(html2, 'lxml')
            soup_texts2 = soup2.find('div', attrs= 'Main List')
            linksa = soup_texts2.find_all('span',attrs='ellipsis')
            linksb = soup_texts2.find_all('span',class_='ellipsis vip')
            f = open('D:/女生/'+link.text+'.txt','w',encoding='utf-8')
            for link in linksa:
                if link not in linksb and '章' in link.text:
                    link = link.parent
                    print(link.text + ':  ', 'http://www.17k.com/'+link.get('href'))
                    download_url = 'http://www.17k.com/'+link.get('href')
                    download_req = urllib.request.Request(download_url, headers = head)
                    download_response = urllib.request.urlopen(download_req)
                    download_html = download_response.read()
                    download_soup = BeautifulSoup(download_html, 'lxml')
                    download_soup_texts = download_soup.find('div', class_ = 'p')
                    f.write(link.text + '\n')
                    f.write(download_soup_texts.text.split('本书首发来自17K小说网')[0])
                    f.write('\n')
            f.close()         
```
