```python
import requests
from bs4 import BeautifulSoup
import jieba
import numpy as np
import re
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def main():
    start_url = r'https://movie.douban.com/subject/1292052/comments?'
    word_sents = []
    for i in range(5):
        try:
            url = start_url + 'start=' + str(20*i)
            r = requests.get(url)
            soup = BeautifulSoup(r.text, 'html.parser')
            contents = soup.find('div', attrs = 'mod-bd')
            sents = contents.find_all('p')
            for sent in sents:
                word_sents.append(sent.text)
        except:
            continue
    return word_sents
  
import os
sents = main()
with open(r'C:\Users\18813\Desktop\0.txt', 'w') as f:
    f.writelines(sents)
    f.close
    
with open(r'C:\Users\18813\Desktop\0.txt') as f:
    txt = f.read()
seg_list = jieba.cut(txt)
clean_seg_list = []
stopwords = ['','','','','','','']
for seg in seg_list:
    if re.search(r'[\u4e00-\u9fa5]+', seg) and seg not in stopwords:
        clean_seg_list.append(seg)
text = ' '.join(clean_seg_list)

pic = np.array(Image.open('xsk.jpg'))
wc = WordCloud(font_path='C:\Windows\Fonts\simsun.ttc',background_color='white', mask=pic)
wc.generate(text)
wc.to_file('xskdjs.jpg')
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import re
from pyecharts import Pie
  
def main(words):
    start_url = r'http://movie.douban.com/top250?'
    word_sents = []
    word_freq = []
    text = []
    for i in range(5):
        try:
            url = start_url + 'start=' + str(25*i)
            r = requests.get(url)
            soup = BeautifulSoup(r.text,'html.parser')
            contents = soup.find('ol',attrs = 'grid_view')
            sents = contents.find_all('p')
            for sent in sents:
                word_sents.append(sent.text)
            string = ''.join(word_sents)
            s = re.findall(r'[\u4e00-\u9fa5]+',string)
            m = re.findall(r'd{4}',string)
            for p in s:
                text.append(p)
            for p in m:
                text.append(p)
        except:
            continue
    for word in words:
        number = text.count(word)
        word_freq.append(number)  
    return word_freq


attr = ['','','','','','','','','','','','','']
v1 = main(attr)
pie = Pie("sorts")
pie.add("", attr, v1, is_label_show=True)
pie.show_config()
pie.render()
```
